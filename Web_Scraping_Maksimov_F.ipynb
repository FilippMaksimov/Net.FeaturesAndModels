```python
import pandas as pd, time, requests, datetime
from bs4 import BeautifulSoup

def habr_search(queries = ['python'], pages = 1):
  url = 'https://habr.com/ru/search/'
  habr_df = pd.DataFrame()

  for query in queries:
    for p in range(1, pages + 1):
      params = {
          'page': p,
          'order': 'relevance',
          'target_type': 'posts',
          'q': query
      }
      response = requests.get(url, params=params)
      soup = BeautifulSoup(response.text)
      articles = soup.find_all('article', 'tm-articles-list__item')

      rows = []
      for article in articles:
        date = article.find('time').get('title')
        try:
          title = article.find('h2', class_ = 'tm-title tm-title_h2').find('a').find('span').text.strip()
          link = 'https://habr.com' + article.find('h2', class_ = 'tm-title tm-title_h2').find('a').get('href')
        except AttributeError:
          title = article.find('h2', class_ = 'tm-megapost-snippet__title').text.strip()
          link = 'https://habr.com' + article.find('a', class_ = 'tm-megapost-snippet__link tm-megapost-snippet__card').get('href')
        finally:
          response = requests.get(link)
          soup = BeautifulSoup(response.text)
          text_tag = soup.find('div', class_ = 'tm-article-body')
          text = text_tag.text.strip() if text_tag else 'Данный блог больше не ведется в Habr'
          rating_tag = soup.find('div', class_ = 'tm-votes-lever tm-article-rating__votes-switcher tm-votes-lever tm-votes-lever_appearance-article')
          total_rating = rating_tag.get('title') if rating_tag else ''
          if total_rating == None: likes = 0
          else:
            start_count = False
            finished = False
            num = ''
            for i in total_rating:
              if finished == True: break
              if i == '↑':
                start_count = True
                continue
              if start_count == True:
                if i != ' ':
                  num += i
                else:
                  finished = True
                  start_count == False
                  break
            likes = num
          time.sleep(0.3)
          rows.append({'Дата': date, 'Заголовок': title, 'Ссылка на материал': link, 'Текст материала': text, 'Количество лайков': likes})

      habr_df = pd.concat([habr_df, pd.DataFrame(rows)], ignore_index=True)
  return habr_df
```

```python
habr_search(queries=['спортивный анализ'], pages=2)
```
